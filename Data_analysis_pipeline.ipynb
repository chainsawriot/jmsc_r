Suppose I want to calculate the X3P/X3PA amoug players with X3PA > 0 and sorted by X3P/X3PA

```r
nbanonzero <- nba[nba$X3PA > 0,]
nbanonzero$X3PP <- nbanonzero$X3P / nbanonzero$X3PA
data.frame(PLAYER = nbanonzero$PLAYER[order(nbanonzero$X3PP, decreasing = TRUE)], 
           X3PP = nbanonzero$X3PP[order(nbanonzero$X3PP, decreasing = TRUE)])
```

## Non-composability

Suppose you need to have also the POS column included, you need to write the whole thing again.

```r
data.frame(PLAYER = nbanonzero$PLAYER[order(nbanonzero$X3PP, decreasing = TRUE)], 
           X3PP = nbanonzero$X3PP[order(nbanonzero$X3PP, decreasing = TRUE)],
           POS = nbanonzero$POS[order(nbanonzero$X3PP, decreasing = TRUE)])
```

Suppose you need to calculate the mean X3PP of players from each POS, you need to create more 'side effect', because the code below didn't work

```r
tapply(data.frame(PLAYER = nbanonzero$PLAYER[order(nbanonzero$X3PP, decreasing = TRUE)], 
           X3PP = nbanonzero$X3PP[order(nbanonzero$X3PP, decreasing = TRUE)],
           POS = nbanonzero$POS[order(nbanonzero$X3PP, decreasing = TRUE)])$X3PP, POS, mean)
```

```r
x3ppdf <- data.frame(PLAYER = nbanonzero$PLAYER[order(nbanonzero$X3PP, decreasing = TRUE)], 
           X3PP = nbanonzero$X3PP[order(nbanonzero$X3PP, decreasing = TRUE)],
           POS = nbanonzero$POS[order(nbanonzero$X3PP, decreasing = TRUE)])
tapply(x3ppdf$X3PP, x3ppdf$POS, mean)
ls() # even more side effect
```

You can eliminate all side-effects using with() to create a temporary namespace, but looking counter-intuitive.

```r
rm(list = ls()) # clear workspace
nba <- read.csv("nba2013.csv", stringsAsFactors = FALSE)

with(nba[nba$X3PA >0,], tapply(X3P/X3PA, POS, mean))

ls()
```

# dplyr

[dplyr](https://github.com/hadley/dplyr) is supposed to be a solution to this by introducing data manipulation verbs.

```r
rm(list = ls()) # clear workspace
require(dplyr)
nba <- read.csv("nba2013.csv", stringsAsFactors = FALSE)
ls()
```

###filter() - selection of rows

```r
filter(nba, X3PA > 0)
```

### mutate() - calculation of new column

```r
mutate(
    filter(nba, X3PA > 0),
    X3PP = X3P / X3PA
)
```

### arrange() - sort by column, use desc() to reverse the order

```r
arrange(
    mutate(
        filter(nba, X3PA > 0),
        X3PP = X3P / X3PA
    ), 
    desc(X3PP)
)
```

### select() - selection of columns

```r
select(
    arrange(mutate(filter(nba, X3PA > 0),X3PP = X3P / X3PA), desc(X3PP)),
    PLAYER, POS, X3P, X3PA, X3PP
)
```

### group_by() - group the data frame by value in column, no effect until further processing.

```r
group_by(
    select(
        arrange(
            mutate(
                filter(nba, X3PA > 0), 
                X3PP = X3P / X3PA
            ), 
            desc(X3PP)
        ),
        PLAYER, POS, X3P, X3PA, X3PP
    ),
    POS
)
```

### summarise() - calculation of summary statistics

and let's look at the the workspace

```r
summarise(
    group_by(
        select(
            arrange(
                mutate(
                    filter(nba, X3PA > 0),
                    X3PP = X3P / X3PA
                ), 
                desc(X3PP)
            ), 
            PLAYER, POS, X3P, X3PA, X3PP
        ), 
        POS
    ), 
    meanX3PP = mean(X3PP)
)
ls()
```

## Readability matters too

It is difficult to read nested parens

Introducing

<img src = "http://itsalocke.com/wp-content/uploads/2015/04/magrittr_m6tn2w.jpg">

%>% - the 'then' operator

Introduced in the magrittr package, make famous by dplyr

```r
nba %>% filter(., X3PA > 0)

# effectively equal to 
# filter(nba, X3PA > 0)

# putting the content in the LHS to the dot
```

If the position of the argument is the first, you can even skip the dot.

```r
nba %>% filter(X3PA > 0)
```

```r
nba %>% filter(X3PA > 0) %>% mutate(X3PP = X3P / X3PA)
```

```r
nba %>% filter(X3PA > 0) %>% mutate(X3PP = X3P / X3PA) %>% 
arrange(desc(X3PP))
# Bias towards those with only one attempts
```

```r
nba %>% filter(X3PA > 0) %>% mutate(X3PP = X3P / X3PA) %>% 
arrange(desc(X3PA), desc(X3PP)) 
# sort by X3PA, then X3PP
```

```r
nba %>% filter(X3PA > 0) %>% mutate(X3PP = X3P / X3PA) %>% 
arrange(desc(X3PA), desc(X3PP)) %>%
select(PLAYER, POS, X3P, X3PA, X3PP)
```

```r
nba %>% filter(X3PA > 0) %>% 
mutate(X3PP = X3P / X3PA) %>% group_by(POS) %>% 
summarise(meanX3PP = mean(X3PP))
```

Advantage - composability

```r
nba %>% filter(X3PA > 0) %>% 
mutate(X3PP = X3P / X3PA) %>% group_by(POS) %>% 
summarise(meanX3PP = mean(X3PP)) %>%
arrange(desc(meanX3PP))
```

```r
require(ggplot2)
nba %>% filter(X3PA > 0) %>% 
mutate(X3PP = X3P / X3PA) %>% group_by(POS) %>% 
summarise(meanX3PP = mean(X3PP)) %>%
arrange(desc(meanX3PP)) %>%
ggplot(aes(x=POS, y=meanX3PP)) + geom_bar(stat = 'identity')
```

# Introducing rio 0.2, the first and last stop of your data analysis pipeline

rio is a collection of functions for reading and writing data from and to data file.

only two functions: import(), export()


```r
require(rio)
import("nba2013.csv") %>% filter(X3P == 0)
```

## On dirtyness

[My comment on chaining circa 2014](https://www.facebook.com/groups/hkrusers/permalink/769863973043766/)

Not always using prefix notation e.g. import("hello.xlsx"), infix is common in some cases.


```r
2 + 3 #infix
'+'(2,3) # prefix
1:10 #infix
':'(1,10) # prefix
```

```r
#same token
mean(rnorm(100)) #prefix
100 %>% rnorm %>% mean #infix, purist form
```

The best practice I guess is not to be purist.
Know when to mix infix notation and prefix notation to make R code more idiomatic and expressive.

```r
100 %>% rnorm %>% mean
# story: create a scaler of 100 then pipe to rnorm and 
# then calculate the mean
# not intuitive

rnorm(100) %>% mean 
# story: generate 100 random numbers 'then' calculate the mean

mean(rnorm(100))
# story: calculate the mean of 100 random numbers
```
